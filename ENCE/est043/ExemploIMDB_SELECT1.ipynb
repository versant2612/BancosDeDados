## Verificar a criação de tabelas em dois esquemas diferentes



O link para a documentação no PostgreSQL é https://www.postgresql.org/docs/current/information-schema.html

Quais esquemas existem no database est043?

```python
%%sql

SELECT DISTINCT table_schema
FROM INFORMATION_SCHEMA.TABLES;
```

```python
%%sql

SELECT *
FROM INFORMATION_SCHEMA.SCHEMATA;
```

Quais são as tabelas dos esquemas cinemav1 e cinemav2?

```python
%%sql

SELECT *
FROM INFORMATION_SCHEMA.TABLES
WHERE table_schema in ('cinemav1', 'cinemav2');
```

```python
%%sql

SELECT table_schema, table_name
FROM INFORMATION_SCHEMA.TABLES
WHERE table_schema LIKE ('cinema%');
```

Quais as colunas da tabela Filmes?

```python
%%sql

SELECT table_schema, table_name, column_name, ordinal_position, is_nullable, data_type
FROM INFORMATION_SCHEMA.COLUMNS
WHERE table_name LIKE ('filme%');
```

```python
%%sql

SELECT table_schema, table_name, column_name, ordinal_position, is_nullable, data_type
FROM INFORMATION_SCHEMA.COLUMNS
WHERE table_name = 'filme';
```

Quais as colunas da tabela Filmes do esquema cinemav1?

```python
%%sql

SELECT table_name, column_name, ordinal_position, is_nullable, data_type
FROM INFORMATION_SCHEMA.COLUMNS
WHERE table_name = 'filme' and table_schema = 'cinemav1';
```

Quais as colunas entre a terceira e a quinta posição nas tabelas Filme dos esquemas de cinema?

```python
%%sql

SELECT table_schema, table_name, column_name, ordinal_position, is_nullable, data_type
FROM INFORMATION_SCHEMA.COLUMNS
WHERE ordinal_position 	BETWEEN 3 AND 5 and table_name = 'filme';
```

Quais as constraints dos esquemas de cinema?


```python
%%sql

SELECT *
FROM INFORMATION_SCHEMA.constraint_column_usage
WHERE table_schema LIKE ('cinema%');
```

Quais as constraint de PK e FK da tabela FilmeElenco do esquema cinemav1?

```python
%%sql

SELECT constraint_name,  column_name,	ordinal_position,	position_in_unique_constraint
FROM INFORMATION_SCHEMA.key_column_usage
WHERE constraint_schema = 'cinemav1' and  table_name = 'filmeelenco';
```

Quais as constraint de Check e Not nNull do esquema cinemav2?

```python
%%sql
SET SCHEMA 'information_schema';
SELECT *
FROM check_constraints
WHERE constraint_schema = 'cinemav2';
```
